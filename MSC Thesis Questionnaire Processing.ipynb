```python
constructs = [competence, imagination,flow,  tension, challenge, negative, positive ]
construct_names = ['Competence', 'Immersion', 'Flow', 'Tension', 'Challenge', 'Negative', 'Positive']

means = []
for i in range(len(constructs)):
    #print(construct_names[i])
    #print(np.mean(post_quest_ng[constructs[i]], axis = 1))
    means.append(np.mean(post_quest_g[constructs[i]], axis = 1))
   
#print(means)
data_gamified = pd.DataFrame(means)
data_gamified = data_gamified.T
data_gamified.columns = construct_names

data_gamified['Condition'] = 'gamified'
print(data_gamified)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import ttest_rel, shapiro, probplot, wilcoxon


# Define the categories
categories = ['Competence', 'Immersion', 'Flow', 'Tension', 'Challenge', 'Negative', 'Positive']

# Conduct paired t-tests and Wilcoxon signed-rank tests, and test assumptions
p_values_ttest = []
p_values_wilcoxon = []
normality_results = []

for category in categories:
    # Calculate the differences
    differences = data_gamified[category] - data_standard[category]
    
    # Paired t-test
    t_stat, p_val_ttest = ttest_rel(data_standard[category], data_gamified[category])
    p_values_ttest.append((category, p_val_ttest))
    
    # Wilcoxon signed-rank test
    w_stat, p_val_wilcoxon = wilcoxon(data_standard[category], data_gamified[category], alternative = 'less')
    p_values_wilcoxon.append((category, p_val_wilcoxon))
    
    # Shapiro-Wilk test for normality
    shapiro_stat, shapiro_p_val = shapiro(differences)
    normality_results.append((category, shapiro_p_val))
    
    # Print results
    print(f'Paired t-test for {category}: p-value = {p_val_ttest}')
    print(f'Wilcoxon signed-rank test for {category}: p-value = {p_val_wilcoxon}')
    print(f'Shapiro-Wilk test for {category} differences: p-value = {shapiro_p_val}')
    
    # Plot Q-Q plot and histogram
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    probplot(differences, dist="norm", plot=plt)
    plt.title(f'Q-Q plot of differences for {category}')
    
    plt.subplot(1, 2, 2)
    plt.hist(differences, bins=20, edgecolor='black')
    plt.title(f'Histogram of differences for {category}')
    plt.xlabel('Difference')
    plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()

# Summary of normality test results
print("\nNormality test p-values (Shapiro-Wilk):")
for category, p_val in normality_results:
    print(f'{category}: {p_val}')

# Summary of test results
print("\nPaired t-test p-values:")
for category, p_val in p_values_ttest:
    print(f'{category}: {p_val}')

print("\nWilcoxon signed-rank test p-values:")
for category, p_val in p_values_wilcoxon:
    print(f'{category}: {p_val}')
```
